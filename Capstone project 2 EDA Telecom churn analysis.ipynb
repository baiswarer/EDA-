# Bivarite Analysis

## 1.Scatter plot

````python
# Box plot for churn coloumn by each numerical feature column
df2 = telecom_df.describe().columns
for col in df2:
  fig = plt.figure(figsize=(17,3))
  ax=fig.gca()
  feature=telecom_df[col]
  label=telecom_df['Churn']
  correlation=feature.corr(label)
  plt.scatter(x=feature,y=label)
  plt.xlabel(col)
  plt.ylabel('Churn')
  plt.title('Scatterplot for churn with other features')
plt.show()
````

Plotted scatter plot which shows the churn vs all the other feature like total night charge/total intl charge etc
